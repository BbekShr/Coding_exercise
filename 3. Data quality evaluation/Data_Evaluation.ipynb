```python
#Import datasets

from google.colab import drive
import pandas as pd

#mount google drive
drive.mount('/content/drive')

# Define file paths
file_paths = {
    "brands": "/content/drive/MyDrive/Colab Notebooks/Data folder/brands.json",
    "receipts": "/content/drive/MyDrive/Colab Notebooks/Data folder/receipts.json",
    "users": "/content/drive/MyDrive/Colab Notebooks/Data folder/users.json",
}

# Create a dictionary to store the DataFrames
dataframes = {}

# Load each file into a DataFrame and store it in the dictionary
for name, path in file_paths.items():
    try:
        # Load the JSON file into a DataFrame
        dataframes[name] = pd.read_json(path, lines=True)
        print(f"{name} DataFrame loaded successfully.")
    except FileNotFoundError:
        print(f"File not found: {path}")
    except Exception as e:
        print(f"An error occurred while loading {name}: {e}")
```

```python
# Brand dataset
brands_df = dataframes.get('brands')
brands_df.head()
```

```python
# receipts dataset
receipts_df = dataframes.get('receipts')
receipts_df.head()
```

```python
# users dataset
users_df = dataframes.get('users')
users_df.head()
```

**Clean and normalize data**

```python
#Users.json file


def convert_timestamp_to_date(df, column_names):
    for column_name in column_names:
        df[column_name] = pd.to_datetime(
            df[column_name].apply(lambda x: x['$date'] if isinstance(x, dict) else x), unit='ms').dt.round('s')
    return df

def unnest_columns(df, column_names):
    for column_name in column_names:
        df[column_name] = df[column_name].apply(lambda x: x['$oid'] if isinstance(x, dict) and '$oid' in x else x)
    return df

columns_to_convert = ['createdDate', 'lastLogin']
users_df = convert_timestamp_to_date(users_df, columns_to_convert)

columns_to_unnest = ['_id']
users_df = unnest_columns(users_df, columns_to_unnest)

# Preview the DataFrame
users_df.head()
```

```python
#brands.json file
columns_to_unnest = ['_id']
brands_df = unnest_columns(brands_df, columns_to_unnest)
brands_df.head()

# Create a new DataFrame with the 'cpg' column from brands_df
cpg_df = brands_df[['cpg']].copy()

# Remove the 'cpg' column from brands_df
brands_df = brands_df.drop(columns=['cpg'])

brands_df.head()
```

**Create cpg.csv**

```python
#Create cpg.csv from brands.json

# Normalize the 'cpg' column
cpg_normalized = pd.json_normalize(cpg_df['cpg'])

# Rename columns
cpg_normalized = cpg_normalized.rename(columns={'$id.$oid': '_id', '$ref': 'ref'})

# Drop the original 'cpg' column and add normalized data
cpg_df = pd.concat([cpg_df.drop(columns=['cpg']), cpg_normalized], axis=1)

# Display the updated DataFrame
cpg_df.head()
```

```python
#receipts.json file

#unnest column
columns_to_unnest = ['_id']
receipts_df = unnest_columns(receipts_df, columns_to_unnest)

columns_to_convert = ['createDate', 'dateScanned', 'finishedDate','modifyDate','pointsAwardedDate','purchaseDate' ]
receipts_df = convert_timestamp_to_date(receipts_df, columns_to_convert)

# Create a new DataFrame with the 'cpg' column from brands_df
rewardsReceiptItemList_df = receipts_df[['_id','rewardsReceiptItemList']].copy()

# Remove the 'cpg' column from brands_df
receipts_df = receipts_df.drop(columns=['rewardsReceiptItemList'])

# Display the first few rows of both DataFrames (optional)
receipts_df.head()
```

```python
rewardsReceiptItemList_df.head()
```

**Create rewardsReceiptItemList.csv**

```python
import pandas as pd
import ast

# Function to safely parse strings into lists of dictionaries
def parse_rewards(value):
    if isinstance(value, str):
        try:
            parsed_value = ast.literal_eval(value)
            # Ensure the output is always a list
            return parsed_value if isinstance(parsed_value, list) else [parsed_value]
        except (ValueError, SyntaxError):
            return []  # Return an empty list if parsing fails
    elif isinstance(value, list):
        return value  # Return as-is if it's already a list
    else:
        return []

# Apply the parsing function to the 'rewardsReceiptItemList' column
rewardsReceiptItemList_df['rewardsReceiptItemList'] = rewardsReceiptItemList_df['rewardsReceiptItemList'].apply(parse_rewards)

expanded_data = []
for idx, row in rewardsReceiptItemList_df.iterrows():
    for item in row['rewardsReceiptItemList']:
        item['_id'] = row['_id']  # Add the '_id' from the original row to each item
        expanded_data.append(item)
normalized_rewards = pd.json_normalize(expanded_data)

# Display the normalized DataFrame with '_id'
normalized_rewards.head()
```

**Explore data qualities issue**


```python
import pandas as pd

# Addint dataframes to a list
dataframes = {
    "users_df": users_df,
    "brands_df": brands_df,
    "cpg_df": cpg_df,
    "receipts_df": receipts_df,
    "normalized_rewards": normalized_rewards
}

# Function to perform basic data quality checks
def evaluate_data_quality(df_name, df):
    print(f"--- Evaluating {df_name} ---")
    results = {}

    # Check for missing values
    results["missing_values"] = df.isnull().sum()

    # Check for duplicate rows
    results["duplicate_rows"] = df.duplicated().sum()

    # Print results
    for key, value in results.items():
        print(f"{key}:")
        print(value)
        print("\n")

# Iterate through all DataFrames in the list
for name, df in dataframes.items():
    evaluate_data_quality(name, df)
```

Dataset


1.	**Users**: There are 283 duplicate values, which seems reasonable since users can log in multiple times. Missing states also appear normal, as some countries do not have states. However, there are a few missing SignUpSource entries. This might be because they belong to returning users, but it would be interesting to investigate why these values are missing, as they could provide insights into user acquisition channels.

2.	**Brands**: There are 234 missing brandCode values. Since brandCode is useful for analyzing how often specific brands appear in receipts, this could impact trend analysis. Alternatively, using barcode as a fallback may help fill the gap in understanding brand trends.

3.	**CPG**: This dataset is derived from Brands, so duplicates shouldn’t be an issue. Most entries seem to have consistent CPG values, which aligns with expectations.

4.	**Receipts**: There are no duplicate rows, which is great. However, some missing values for purchasedItemCount and purchaseDate limit the insights we can gain from the data. These values could potentially be extracted from each receipt to provide a more comprehensive view of purchasing trends.

5.	**RewardsReceiptItemList** (normalized_rewards): While it’s expected to have some missing rows, the absence of critical data such as brandCode, barcode, and item price reduces the ability to analyze user trends effectively. This information is vital for understanding which brands users are buying and how much they are spending.


```python
# List of columns that can be null but not negative
columns_to_check = ["bonusPointsEarned", "purchasedItemCount", "finalPrice", "itemPrice", "targetPrice"]

# Function to check for negative values
def validate_no_negative_values(df, columns):
    results = {}
    for column in columns:
        if column in df.columns:
            # Convert to numeric and handle non-numeric values as NaN
            df[column] = pd.to_numeric(df[column], errors='coerce')

            # Count negative values, ignoring NaN values
            negative_count = (df[column] < 0).sum()
            results[column] = negative_count
    return results

# Iterate through all DataFrames in the dictionary
for name, df in dataframes.items():
    print(f"Evaluating data quality for: {name}")
    negative_value_issues = validate_no_negative_values(df, columns_to_check)
    print("Summary of negative value issues:")
    print(negative_value_issues)
```

Check for columns such as ["bonusPointsEarned", "purchasedItemCount", "finalPrice", "itemPrice", "targetPrice"] that cannot be less than zero. All columns passed the test
